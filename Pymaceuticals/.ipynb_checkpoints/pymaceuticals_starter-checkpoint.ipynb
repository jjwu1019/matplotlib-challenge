```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_pd = pd.read_csv(mouse_drug_data_to_load)
clinical_pd = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_table = pd.merge(clinical_pd, mouse_drug_pd, on="Mouse ID")

# Display the data table for preview
merge_table
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
merge_table_temp = merge_table[['Drug', 'Timepoint', 'Tumor Volume (mm3)']]
merge_table_groupby_drug = merge_table_temp.groupby(['Drug', 'Timepoint'])

# Convert to DataFrame
merge_table_groupby_drug_mean = merge_table_groupby_drug.mean()
merge_table_groupby_drug_mean = merge_table_groupby_drug_mean.reset_index()
# Preview DataFrame
merge_table_groupby_drug_mean
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
merge_table_groupby_drug_sem = merge_table_groupby_drug.sem()
# Convert to DataFrame
merge_table_groupby_drug_sem = merge_table_groupby_drug_sem.reset_index()
# Preview DataFrame
merge_table_groupby_drug_sem.head(5)
```

```python
# Minor data mung to reformat standard error for my graph function
merge_table_groupby_drug_sem_mung = merge_table_groupby_drug_sem.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
merge_table_groupby_drug_sem_mung
```

```python
# Minor Data Munging to Re-Format the Data Frames
merge_table_groupby_drug_mean_mung = merge_table_groupby_drug_mean.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
merge_table_groupby_drug_mean_mung
```

```python
# the function I call for graphing, where data is the dataframe after munging, and sem is the standard error dataframe after munging
def draw_graph(data, sem):
    col_names = list(data.columns)
    for x in range(len(col_names)):
        values = data.iloc[:, x]
        plt.errorbar(values.index, values, yerr = sem.iloc[:,x], label=col_names[x], marker="o", linewidth=0.7, markersize=3)
    plt.legend()
```

```python
# Generate the Plot (with Error Bars)
# draw_graph is the function I made, read above for detail
draw_graph(merge_table_groupby_drug_mean_mung, merge_table_groupby_drug_sem_mung)

plt.title("Tumor Response to treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(list(merge_table_groupby_drug_mean_mung.columns), loc="best")
plt.grid()

plt.xlim(-0.75, 70)
plt.ylim(30, 75)

# Save the Figure
plt.savefig("treatment.png")
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
merge_table_temp_meta = merge_table[['Drug', 'Timepoint', 'Metastatic Sites']]
merge_table_groupby_drug_meta = merge_table_temp_meta.groupby(['Drug', 'Timepoint'])

# Convert to DataFrame
merge_table_groupby_drug_mean_meta = merge_table_groupby_drug_meta.mean()
merge_table_groupby_drug_mean_meta = merge_table_groupby_drug_mean_meta.reset_index()
# Preview DataFrame
merge_table_groupby_drug_mean_meta
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
merge_table_groupby_drug_sem_meta = merge_table_groupby_drug_meta.sem()
# Convert to DataFrame
merge_table_groupby_drug_sem_meta = merge_table_groupby_drug_sem_meta.reset_index()
# Preview DataFrame
merge_table_groupby_drug_sem_meta.head(5)
```

```python
# Minor data mung to reformat standard error for my graph function
merge_table_groupby_drug_sem_meta_mung = merge_table_groupby_drug_sem_meta.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
merge_table_groupby_drug_sem_meta_mung
```

```python
# Minor Data Munging to Re-Format the Data Frames
merge_table_groupby_drug_mean_mung_meta = merge_table_groupby_drug_mean_meta.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
# Preview that Reformatting worked
merge_table_groupby_drug_mean_mung_meta
```

```python
# Generate the Plot (with Error Bars)
draw_graph(merge_table_groupby_drug_mean_mung_meta,merge_table_groupby_drug_sem_meta_mung)

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.legend(list(merge_table_groupby_drug_mean_mung_meta.columns), loc="best")
plt.grid()

plt.xlim(-0.75, 70)
plt.ylim(-0.1, 3.7)

# Save the Figure
plt.savefig("spread.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
merge_table_temp = merge_table[['Drug', 'Timepoint', 'Tumor Volume (mm3)']]
merge_table_groupby_drug = merge_table_temp.groupby(['Drug', 'Timepoint'])

# Convert to DataFrame
merge_table_groupby_drug_count = merge_table_groupby_drug.count()
merge_table_groupby_drug_count = merge_table_groupby_drug_count.reset_index()
# Preview DataFrame
merge_table_groupby_drug_count = merge_table_groupby_drug_count.rename(columns={"Tumor Volume (mm3)": "Mouse Count"})
merge_table_groupby_drug_count.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
merge_table_groupby_drug_count_mung = merge_table_groupby_drug_count.pivot(index='Timepoint', columns='Drug', values='Mouse Count')
# Preview the Data Frame
merge_table_groupby_drug_count_mung
```

```python
# Generate the Plot (Accounting for percentages)
merge_table_groupby_drug_count_mung.loc[5] = merge_table_groupby_drug_count_mung.loc[5]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[10] = merge_table_groupby_drug_count_mung.loc[10]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[15] = merge_table_groupby_drug_count_mung.loc[15]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[20] = merge_table_groupby_drug_count_mung.loc[20]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[25] = merge_table_groupby_drug_count_mung.loc[25]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[30] = merge_table_groupby_drug_count_mung.loc[30]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[35] = merge_table_groupby_drug_count_mung.loc[35]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[40] = merge_table_groupby_drug_count_mung.loc[40]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[45] = merge_table_groupby_drug_count_mung.loc[45]/merge_table_groupby_drug_count_mung.loc[0]*100
merge_table_groupby_drug_count_mung.loc[0] = merge_table_groupby_drug_count_mung.loc[0]/merge_table_groupby_drug_count_mung.loc[0]*100


survival_plot = plt.plot(merge_table_groupby_drug_count_mung, marker="o")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate(%)")
plt.legend(list(merge_table_groupby_drug_count_mung), loc="best")
plt.grid()

plt.xlim(-0.75, 70)
plt.ylim(20, 101)
# Save the Figure
plt.savefig("survival.png")
# Show the Figure
plt.show()
```
